```python
# Assign Food Categories

mos["Food_Category"] = (
    mos["Item_Code"]
    .map(food_mapping)
)

# Keep Food Records 

mos = mos[
    mos["Food_Category"].notna()
].copy()

print(mos.shape)
```

```python
# Create Household Food Matrix

food_hh = (
    mos.pivot_table(
        index=[
            "HH_ID",
            "State",
            "District",
            "Sector"
        ],
        columns="Food_Category",
        values="Total_Consumption_Value",
        aggfunc="sum",
        fill_value=0
    )
    .reset_index()
)
```

```python
# Ensure All Food Columns Exist

food_cols = [
    "Cereals",
    "Pulses",
    "Milk_Products",
    "Vegetables",
    "Fruits",
    "Egg_Fish_Meat",
    "Edible_Oil",
    "Spices",
    "Beverages",
    "Sugar"
]

for col in food_cols:
    if col not in food_hh.columns:
        food_hh[col] = 0
```

```python
# Add Household Weights

hh_weight = (
    mos.groupby("HH_ID")["Multiplier"]
    .first()
    .reset_index()
)

food_hh = food_hh.merge(
    hh_weight,
    on="HH_ID",
    how="left"
)
```

```python
# Calculate Total Food Expenditure

food_hh["Total_Food_Value"] = (
    food_hh[food_cols]
    .sum(axis=1)
)

# Calculate Out-of-Home Food Expenditure

ooh = (
    mos.groupby("HH_ID")
    ["OutOfHome_Consumption_Value"]
    .sum()
    .reset_index()
)

ooh.rename(
    columns={
        "OutOfHome_Consumption_Value":
        "OutOfHome_Food_Value"
    },
    inplace=True
)
```

```python
# Merge Out-of-Home Data

food_hh = food_hh.merge(
    ooh,
    on="HH_ID",
    how="left"
)

food_hh["OutOfHome_Food_Value"] = (
    food_hh["OutOfHome_Food_Value"]
    .fillna(0)
)
```

```python
# Food Diversity Score
food_hh["Food_Items_Count"] = (
    food_hh[food_cols] > 0
).sum(axis=1)

food_hh["Diet_Diversity_Score"] = (
    food_hh["Food_Items_Count"] / 10
)

# Nutrition Score
nutrition_cols = [
    "Milk_Products",
    "Fruits",
    "Vegetables",
    "Egg_Fish_Meat",
    "Pulses"
]

food_hh["Nutrition_Score"] = (
    (
        food_hh[nutrition_cols] > 0
    ).sum(axis=1)
    / len(nutrition_cols)
)

# Protein Score
food_hh["Protein_Value"] = (
    food_hh["Pulses"]
    + food_hh["Milk_Products"]
    + food_hh["Egg_Fish_Meat"]
)

food_hh["Protein_Score"] = (
    food_hh["Protein_Value"]
    .rank(pct=True)
)

# Food Security Score
food_hh["Food_Security_Score"] = (
    food_hh["Total_Food_Value"]
    .rank(pct=True)
)
```

```python
# Out-of-Home Ratio

food_hh["OutOfHome_Ratio"] = np.where(
    food_hh["Total_Food_Value"] > 0,
    food_hh["OutOfHome_Food_Value"]
    /
    food_hh["Total_Food_Value"],
    0
)

food_hh["OutOfHome_Ratio"] = (
    food_hh["OutOfHome_Ratio"]
    .clip(upper=1)
)
```

```python
# Overall Food Wellbeing Index

food_hh["Overall_Food_Wellbeing_Index"] = (
      0.30 * food_hh["Diet_Diversity_Score"]
    + 0.25 * food_hh["Nutrition_Score"]
    + 0.25 * food_hh["Food_Security_Score"]
    + 0.20 * food_hh["Protein_Score"]
)
```

```python
# Assign Welfare Categories

lv9["Welfare_Category"] = (
    lv9["Item_Code_9_1_to_11_4"]
    .map(welfare_map)
)

# Create Household Welfare Matrix

welfare_hh = (
    lv9.pivot_table(
        index=[
            "HH_ID",
            "State",
            "District",
            "Sector"
        ],
        columns="Welfare_Category",
        values="Value_Rs_9_1_to_11_4",
        aggfunc="sum",
        fill_value=0
    )
    .reset_index()
)
```

```python
# Ensure All Welfare Columns Exist

welfare_cols = [
    "Education",
    "Health",
    "Hygiene",
    "Transport",
    "Housing"
]

for col in welfare_cols:
    if col not in welfare_hh.columns:
        welfare_hh[col] = 0
```

```python
# Merge Household Weights
welfare_hh = welfare_hh.merge(
    hh_weight_lv9,
    on="HH_ID",
    how="left"
)

# Total Welfare Expenditure
welfare_hh["Total_Welfare_Expense"] = (
    welfare_hh[welfare_cols]
    .sum(axis=1)
)

# Welfare Category Count
welfare_hh["Welfare_Category_Count"] = (
    (
        welfare_hh[welfare_cols] > 0
    ).sum(axis=1)
)

# Welfare Diversity Score
welfare_hh["Welfare_Diversity_Score"] = (
    welfare_hh["Welfare_Category_Count"]
    / len(welfare_cols)
)

# Welfare Spending Score
welfare_hh["Welfare_Spending_Score"] = (
    welfare_hh["Total_Welfare_Expense"]
    .rank(pct=True)
)

# Welfare Index
welfare_hh["Welfare_Index"] = (
      0.60 * welfare_hh["Welfare_Spending_Score"]
    + 0.40 * welfare_hh["Welfare_Diversity_Score"]
)
```

```python
# Final Dataset

final_cols = [

    "HH_ID",
    "State",
    "District",
    "Sector",
    "Multiplier",

    "Personal_Goods",
    "Transport_Assets",
    "Sports_Assets",
    "Medical_Equipment",
    "Household_Appliances",
    "Furniture_Fixtures",
    "Residential_Durables",
    "Jewellery_Assets",

    "Total_Durable_Expenditure",
    "Asset_Category_Count",
    "Asset_Diversity_Score",
    "Asset_Wealth_Score",
    "Durable_Asset_Index",
    "Log_Asset_Value"
]

asset_hh = asset_hh[final_cols]
```

### Combining Multiple Data Levels for Comprehensive Analysis

```python
print('MOSPI Level-5 Dataset Shape',food_hh.shape)
print('MOSPI Level-9 Dataset Shape',welfare_hh.shape)
print('MOSPI Level-13 Dataset Shape',asset_hh.shape)
```

```python
# Find Common Households

common_hh = (
    set(food_hh["HH_ID"])
    &
    set(welfare_hh["HH_ID"])
    &
    set(asset_hh["HH_ID"])
)

print("Common HH:", len(common_hh))
```

```python
# Keep Only Common Households

food_final = food_hh[
    food_hh["HH_ID"].isin(common_hh)
].copy()

welfare_final = welfare_hh[
    welfare_hh["HH_ID"].isin(common_hh)
].copy()

asset_final = asset_hh[
    asset_hh["HH_ID"].isin(common_hh)
].copy()
```

```python
# Verify Multipliers Match

mult_check = (
    food_final[["HH_ID", "Multiplier"]]
    .rename(columns={"Multiplier":"Food_Mult"})
    .merge(
        welfare_final[["HH_ID","Multiplier"]]
        .rename(columns={"Multiplier":"Welfare_Mult"}),
        on="HH_ID"
    )
    .merge(
        asset_final[["HH_ID","Multiplier"]]
        .rename(columns={"Multiplier":"Asset_Mult"}),
        on="HH_ID"
    )
)
```

```python
# Remove Bad Households

bad_hh = mult_check[
    (mult_check["Food_Mult"] != mult_check["Welfare_Mult"])
    |
    (mult_check["Food_Mult"] != mult_check["Asset_Mult"])
]["HH_ID"]

print("Bad HH:", len(bad_hh))
```

```python
# Remove Them

food_final = food_final[
    ~food_final["HH_ID"].isin(bad_hh)
]

welfare_final = welfare_final[
    ~welfare_final["HH_ID"].isin(bad_hh)
]

asset_final = asset_final[
    ~asset_final["HH_ID"].isin(bad_hh)
]
```

```python
# Merge Datasets

master = (
    food_final
    .merge(
        welfare_final.drop(
            columns=[
                "State",
                "District",
                "Sector",
                "Multiplier"
            ]
        ),
        on="HH_ID",
        how="inner"
    )
    .merge(
        asset_final.drop(
            columns=[
                "State",
                "District",
                "Sector",
                "Multiplier"
            ]
        ),
        on="HH_ID",
        how="inner"
    )
)

print(master.shape)
```

```python
# Create Welfare Index

master["Welfare_Index"] = (
    0.5 *
    master["Welfare_Diversity_Score"]
    +
    0.5 *
    master["Total_Welfare_Expense"].rank(pct=True)
)
```

```python
# Normalize All Three Pillars

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

master[
    [
        "Food_Index_N",
        "Welfare_Index_N",
        "Asset_Index_N"
    ]
] = scaler.fit_transform(
    master[
        [
            "Overall_Food_Wellbeing_Index",
            "Welfare_Index",
            "Durable_Asset_Index"
        ]
    ]
)
```

```python
# Create Composite Wellbeing Index

master["Economic_Wellbeing_Index"] = (
      0.40 * master["Food_Index_N"]
    + 0.30 * master["Welfare_Index_N"]
    + 0.30 * master["Asset_Index_N"]
)

# Create Wellbeing Classes
master["Wellbeing_Class"] = pd.qcut(
    master["Economic_Wellbeing_Index"],
    q=5,
    labels=[
        "Very Low",
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
)
```

        District codes for India.
            
        It is used to ensure consistent spatial alignment and merging across MOSPI Level-5, Level-9, Level-13, and 
        external geospatial datasets.
            
        In this project, it enables accurate district-level aggregation and mapping of socio-economic indicators.
            
        This dataset is essential for building a unified geospatial framework for spatial inequality and hotspot analysis.

```python
# Create State Mapping

state_map = (
    dist[
        ["State code", "State Name"]
    ]
    .drop_duplicates()
)

state_map.columns = [
    "State",
    "State_Name"
]
```

```python
# Create District Mapping

district_map_lookup = (
    dist[
        [
            "State code",
            "District code",
            "District Name"
        ]
    ]
    .drop_duplicates()
)

district_map_lookup.columns = [
    "State",
    "District",
    "District_Name"
]
```

### CPI (Consumer Price Index) Dataset

        The CPI dataset provides temporal price level data across states, sectors, and consumption groups, capturing 
        inflation dynamics in India.
            
        It is used to adjust nominal values into real economic terms, ensuring comparability of consumption and welfare
        indicators over time.
            
        In this project, it supports inflation correction for MOSPI consumption and expenditure variables to build 
        accurate wellbeing measures.
            
        This dataset is essential for integrating time-aware economic stability into spatial inequality and welfare modeling.

```python
cpi = pd.read_excel("cpi_1061.xlsx")
print(cpi.shape)
```

```python
cpi.head(5)
```

```python
cpi.isnull().sum()
```

```python
# Keep only 2024

cpi = cpi[
    cpi["year"] == 2024
].copy()

print(cpi.shape)
print(cpi["year"].unique())
```

```python
# Prepare State Names in Master

master["cpi_state"] = (
    master["State_Name"]
    .str.strip()
)

master["cpi_state"] = master["cpi_state"].replace({
    "Andaman & Nicobar Island": "Andaman & Nicobar Islands",
    "Ladakh": "Jammu & Kashmir"
})
```

```python
# Verify State Matching

master_states = set(
    master["cpi_state"].unique()
)

cpi_states = set(
    cpi["state"].str.strip().unique()
)

print("MASTER NOT IN CPI")
print(master_states - cpi_states)

print("\nCPI NOT IN MASTER")
print(cpi_states - master_states)
```

```python
# Create CPI Indicators
# Food CPI

food_cpi = (
    cpi[
        cpi["group"] == "Food and Beverages"
    ]
    .groupby("state")["index"]
    .mean()
    .reset_index()
)

food_cpi.columns = [
    "state",
    "Food_CPI"
]
```

```python
# Housing CPI

housing_cpi = (
    cpi[
        cpi["group"] == "Housing"
    ]
    .groupby("state")["index"]
    .mean()
    .reset_index()
)

housing_cpi.columns = [
    "state",
    "Housing_CPI"
]
```

```python
# Health CPI

health_cpi = (
    cpi[
        cpi["subgroup"] == "Health"
    ]
    .groupby("state")["index"]
    .mean()
    .reset_index()
)

health_cpi.columns = [
    "state",
    "Health_CPI"
]
```

```python
# Education CPI

education_cpi = (
    cpi[
        cpi["subgroup"] == "Education"
    ]
    .groupby("state")["index"]
    .mean()
    .reset_index()
)

education_cpi.columns = [
    "state",
    "Education_CPI"
]
```

```python
# Transport CPI

transport_cpi = (
    cpi[
        cpi["subgroup"] == "Transport and Communication"
    ]
    .groupby("state")["index"]
    .mean()
    .reset_index()
)

transport_cpi.columns = [
    "state",
    "Transport_CPI"
]

# Fuel CPI
fuel_cpi = (
    cpi[
        cpi["group"] == "Fuel and Light"
    ]
    .groupby("state")["index"]
    .mean()
    .reset_index()
)

fuel_cpi.columns = [
    "state",
    "Fuel_CPI"
]
```

```python
# General CPI

general_cpi = (
    cpi[
        cpi["group"] == "General"
    ]
    .groupby("state")["index"]
    .mean()
    .reset_index()
)

general_cpi.columns = [
    "state",
    "General_CPI"
]

# Misc CPI

misc_cpi = (
    cpi[
        cpi["group"] == "Miscellaneous"
    ]
    .groupby("state")["index"]
    .mean()
    .reset_index()
)

misc_cpi.columns = [
    "state",
    "Misc_CPI"
]
```

```python
# Create Inflation Indicators

food_inf = (
    cpi[
        cpi["group"] == "Food and Beverages"
    ]
    .groupby("state")["inflation"]
    .mean()
    .reset_index()
)

food_inf.columns = [
    "state",
    "Food_Inflation"
]

# General Inflation

general_inf = (
    cpi[
        cpi["group"] == "General"
    ]
    .groupby("state")["inflation"]
    .mean()
    .reset_index()
)

general_inf.columns = [
    "state",
    "General_Inflation"
]
```

```python
# Build Final CPI Dataset

cpi_final = (
    food_cpi
    .merge(housing_cpi,
           on="state",
           how="outer")
    .merge(health_cpi,
           on="state",
           how="outer")
    .merge(education_cpi,
           on="state",
           how="outer")
    .merge(transport_cpi,
           on="state",
           how="outer")
    .merge(fuel_cpi,
           on="state",
           how="outer")
    .merge(general_cpi,
           on="state",
           how="outer")
    .merge(misc_cpi,
           on="state",
           how="outer")
    .merge(food_inf,
           on="state",
           how="left")
    .merge(general_inf,
           on="state",
           how="left")
)

print(cpi_final.shape)
cpi_final.head()
```

```python
cpi_final.isnull().sum()
```

```python
# Merge CPI Into Master

master = master.merge(
    cpi_final,
    left_on="cpi_state",
    right_on="state",
    how="left"
)

print(master.shape)
```

```python
# Set display option to show all rows
pd.set_option('display.max_rows', None)

# Now run your check
print(master.isnull().sum())

# Optional: Reset the display option back to default afterwards
pd.reset_option('display.max_rows')
```

```python
# Handle Missing CPI Values

for col in [
    "Health_CPI",
    "Education_CPI",
    "Transport_CPI"
]:
    master[col] = master[col].fillna(
        master["Misc_CPI"]
    )

housing_median = (
    master["Housing_CPI"]
    .median()
)

master["Housing_CPI"] = (
    master["Housing_CPI"]
    .fillna(housing_median)
)
```

```python
# Check Missing Values

cpi_cols = [
    "Food_CPI",
    "Housing_CPI",
    "Health_CPI",
    "Education_CPI",
    "Transport_CPI",
    "Fuel_CPI",
    "General_CPI",
    "Food_Inflation",
    "General_Inflation"
]

print(
    master[cpi_cols]
    .isnull()
    .sum()
)
```

```python
# Create Normalized CPI Features

for col in [
    "Food_CPI",
    "Housing_CPI",
    "Health_CPI",
    "Education_CPI",
    "Transport_CPI",
    "Fuel_CPI",
    "General_CPI"
]:
    master[col + "_N"] = (
        master[col]
        /
        master[col].mean()
    )
```

```python
# Create Real (Inflation Adjusted) Indicators

master["Housing_Burden"] = (
    master["Housing"]
    /
    master["Housing_CPI"]
)

master["Real_Food_Index"] = (
    master["Food_Index_N"]
    /
    master["Food_CPI_N"]
)

master["Real_Housing_Index"] = (
    master["Housing_Burden"]
    /
    master["Housing_CPI_N"]
)

master["Real_Wellbeing"] = (
    (
        master["Food_Index_N"]
        /
        master["Food_CPI_N"]
    )
    +
    (
        master["Welfare_Index_N"]
        /
        master["General_CPI_N"]
    )
    +
    master["Asset_Index_N"]
) / 3
```

```python
# Final Validation

print(master.shape)

print(
    master.isnull().sum()[
        master.isnull().sum() > 0
    ]
)

master[
    [
        "Real_Wellbeing",
        "Food_Inflation",
        "General_Inflation"
    ]
].describe()
```

```python
master.shape
```

```python
# Log-scaled income proxy stability

master["Log_Food_Value"] = np.log1p(master["Total_Food_Value"])
master["Log_Welfare_Expense"] = np.log1p(master["Total_Welfare_Expense"])
master["Log_Asset_Value_Final"] = np.log1p(master["Total_Durable_Expenditure"])

# Inequality / imbalance indicator

master["Consumption_Imbalance"] = (
    master["Food_Index_N"]
    - master["Asset_Index_N"]
).abs()

# Poverty-risk proxy (important for research)

master["Low_Wellbeing_Flag"] = (
    master["Economic_Wellbeing_Index"] < 0.4
).astype(int)

# High-end consumption flag

master["Luxury_Household_Flag"] = (
    master["Jewellery_Assets"] > master["Jewellery_Assets"].quantile(0.90)
).astype(int)

# Rural–Urban gap feature

master["Urban_Premium"] = (
    master.groupby("Sector_Name")["Economic_Wellbeing_Index"]
    .transform("mean")
)

# Composite vulnerability index

master["Vulnerability_Index"] = (
    1 - master["Economic_Wellbeing_Index"]
) * (1 + master["Housing_Burden"])
```

### Final Integrated Analytical Dataset (EDA & Machine Learning)

        The final analytical dataset was created by integrating MOSPI HCES Level-5, Level-9, Level-13, district administrative 
        data, and Consumer Price Index (CPI) information into a unified household-level database. 
        
        The integrated dataset contains household identifiers, socio-economic indicators, inflation-adjusted variables, and 
        engineered features designed for exploratory data analysis (EDA), spatial analysis, and machine learning.

### The dataset is organized into the following feature groups:

        Household Identifier:- Household ID (HH_ID).
        
        Food Consumption Features:- Food expenditure by category (Cereals, Pulses, Milk Products, Fruits, Vegetables, Edible Oil, 
                                    Sugar, Spices, Beverages, Egg/Fish/Meat), total food expenditure, dietary diversity, 
                                    nutrition score, protein score, food security score, out-of-home food expenditure, and 
                                    Overall Food Wellbeing Index.
        
        Welfare Features:- Expenditure on Education, Health, Hygiene, Housing, and Transport, together with Welfare Diversity 
                           Score, Welfare Spending Score, and Welfare Index.
                           
        Household Asset Features:- Household durable assets including transport assets, household appliances, furniture, 
                                   residential durables, jewellery, medical equipment, sports assets, and personal goods, 
                                   summarized through Asset Diversity Score, Asset Wealth Score, Durable Asset Index, 
                                   and logarithmic asset value.
                                   
        Economic Wellbeing Indicators:- Normalized Food, Welfare, and Asset indices combined into the Economic Wellbeing Index 
                                        and household Wellbeing Class.
                                        
        Administrative and Spatial Variables:- State, district, sector (Rural/Urban), and household multiplier used for spatial 
                                               aggregation and district-level analysis.
                                               
        Macroeconomic Variables:- State-level Consumer Price Index (Food, Housing, Health, Education, Transport, Fuel, General, 
                                    and Miscellaneous CPI) and corresponding inflation indicators.
                                  
        Inflation-Adjusted Indicators:- Real Food Index, Real Housing Index, and Real Wellbeing, constructed using CPI-adjusted 
                                        expenditure measures.
                                        
        Machine Learning Features:- Engineered variables including Housing Burden, Consumption Imbalance, Vulnerability Index, 
                                    Urban Premium, Luxury Household Flag, Low Wellbeing Flag, and logarithmic transformations of 
                                    expenditure variables.

        The final integrated dataset provides a multidimensional representation of household consumption, welfare, 
        asset ownership, and inflation-adjusted economic conditions, forming the foundation for exploratory data analysis,
        spatial hotspot detection, econometric modeling, and machine learning-based wellbeing prediction.
